# DISCOVER Hack Shack Challenges

Provided by [HPEDEV Team](hpedev.io)

# Challenge process:

Every HPE Discover Virtual Experience 2020 Hack Shack challenge uses GitHub and Jupyter Notebooks to provide instructions for the challenge and to accept the answers submitted by participants.

## Getting started 

- Step 1: [Register](URL) for the challenge to obtain your Jupyter account 
- Step 2: Fork the challenge you'd like to take from https://github.com/HPEDevCom
- Step 3: Clone your copy of the challenge into your Jupyter environment (using your Launcher terminal window)
- Step 4: open challenge notebook in your Jupyter environment and start working on it

Some of the challenges have Workshop prerequisites. The notebooks of these workshops have been copied in your Jupyter environment. Don't hesitate to refer to those notebooks if you missed the live workshop. The GIT101 Workshop is a prerequisite for all challenges and its content can also be found in your Jupyter environment.

Please note, the account will only be allocated for a 4-hour time slot, so make sure you register only when you have enough time available to take the challenge.

## Selecting your challenge

We will offer the following four challenges over the course of this HPE Discover Virtual Experience:

1. The OneView Challenge
2. The Redfish Challenge
3. The Grommet Challenge
4. The Container Challenge

If time permits, we encourage you to take all of them in this order and if the subject is relevant to your job.

### The Redfish Challenge 

In this challenge we will ask you to write the smallest possible program (in terms of number of Redfish API calls), to retrieve the complete Device Inventory list from an HPE iLO 5 and display the following properties for each device: Location, Product Name, Firmware Version (if applicable) and Status. Be quick and be sharp!

Estimated time to complete: 1 hour

### The Grommet Challenge 

In this challenge, we will ask to be creative and show the UX designer in you. You will need to use Grommet to design your own little web app UI. You will start with the Grommet Designer, generate code from your design, push that code to GitHub and finally deploy the app on Netlify. Beginner, expert designers and developers are all welcome. It's all about unleashing your creativity!

Estimated time to complete: 2 hours

### The Container Challenge 

This challenge is undoubtly the most complicated one as it's a end-to-end developer experience. In this challenge you will be asked to build a Grommet sample app (or reuse the one from the Grommet Challenge), package it in a container, then run that container in a Kubernetes Cluster. This challenge touches on several technologies such as front-end web application design, Docker containers and Kubernetes. Be ready for an intense moment in which all the pieces of the puzzle will suddenly make total sense.

Estimated time to complete: 4 hours

## Submitting your response

When you are ready to submit your challenge notebook, we will ask you to leverage GitHub again to do the following:

- Step 1: Commit changes in your Jupyter environment  (using your Launcher terminal window)
- Step 2: Push changes to your repo  (using your Launcher terminal window)
- Step 3: Open a pull request on our original repo (use the description to tell us your email, so we can contact you)

You should submit your response within the 4-hour time slot. We will reset all the Jupyter accounts right after the time is up, so make sure you have committed and pushed your changes back to your repo prior to that. You then have more time to submit your PR.
